```python
# List of columns to remove
columns_to_remove = ['OBJECTID', 'FIREOCCURID', 'CN', 'REVDATE', 'COMPLEXNAME', 
                     'UNIQFIREID', 'SOFIRENUM', 'LOCALFIRENUM', 'COMMENTS',
                     'UNITIDPROTECT', 'LATDD83', 'LONGDD83',
                     'FIRETYPECATEGORY', 'DBSOURCEID', 'DBSOURCEDATE']

# Remove the redundant columns
df.drop(columns=columns_to_remove, inplace=True)
```

```python
# Rename the X and Y columns as LONGITUDE and LATITUDE
df = df.rename(columns={'X': 'LONGITUDE', 'Y': 'LATITUDE'})
```

```python
# Clean the FIREYEAR column by keeping only years within a reasonable range
df = df[(df['FIREYEAR'] > 1900) & (df['FIREYEAR'] < 2100)]

# min and max function
min_year = df['FIREYEAR'].min()
max_year = df['FIREYEAR'].max()

print(f"The range of FIREYEAR is from {min_year} to {max_year}")
```

```python
# Set our threshold as at least 100 acres
threshold = 100

# Filter out small fires
df = df[df['TOTALACRES'] > threshold]
```

```python
# Display the number of rows in the DataFrame
number_of_rows = len(df)
print("Number of rows:", number_of_rows)
```
