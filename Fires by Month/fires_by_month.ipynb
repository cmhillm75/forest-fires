```python
fire_df_ready = fire_df[['GLOBALID', 'FIRENAME','FIREYEAR','STATCAUSE','TOTALACRES','LATITUDE','LONGITUDE',
       'DISCOVERY_DATE','FIREOUT_DATE','MONTH_NAME']]
```

```python
fire_df_ready.loc[:, 'STATCAUSE'] = fire_df_ready['STATCAUSE'].fillna('Undetermined')
```

```python
fire_df_ready.info()
```

```python
fire_df_ready.loc[:, 'DISCOVERY_DATE'] = pd.to_datetime(fire_df_ready['DISCOVERY_DATE'], errors='coerce')
fire_df_ready.loc[:, 'FIREOUT_DATE'] = pd.to_datetime(fire_df_ready['FIREOUT_DATE'], errors='coerce')
```

```python
fire_df_ready.loc[:, 'DISCOVERY_DATE'] = fire_df_ready['DISCOVERY_DATE'].dt.date
fire_df_ready.loc[:, 'FIREOUT_DATE'] = fire_df_ready['FIREOUT_DATE'].dt.date
```

```python
fire_df_ready.info()
```

```python
fire_df_ready.loc[:,'MONTH_NUMBER'] = fire_df_ready['DISCOVERY_DATE'].dt.month
```

```python
fire_df_ready.info()
```

```python
print(fire_df_ready.columns)
```

```python
fire_df_final = fire_df_ready[['GLOBALID', 'FIRENAME', 'FIREYEAR', 'STATCAUSE', 'TOTALACRES',
       'LATITUDE', 'LONGITUDE', 'DISCOVERY_DATE', 'FIREOUT_DATE', 'MONTH_NUMBER','MONTH_NAME'
       ]]
```

```python
fire_df_final.head()
```

```python
fire_df_final.loc[:, 'DECADE'] = fire_df_final['FIREYEAR'].astype(int).astype(str).str[:3] + "0s"
```

```python
fire_df_final.head()
```

```python
# Define size class bins and labels
bins = [100, 299.99, 999.99, 4999.99, 9999.99, 49999.99, 99999.99, 499999.99, 999999.99, float('inf')]
labels = [
    'Class D',  # 100 - 299.99 acres
    'Class E',  # 300 - 999.99 acres
    'Class F',  # 1000 - 4999.99 acres
    'Class G',  # 5000 - 9999.99 acres
    'Class H',  # 10,000 - 49999.99 acres
    'Class I',  # 50,000 - 99999.99 acres
    'Class J',  # 100,000 - 499999.99 acres
    'Class K',  # 500,000 - 999999.99 acres
    'Class L'   # 1,000,000+ acres
]

# Apply pd.cut() to create the new column
fire_df_ready.loc[:, 'SIZE_CLASS'] = pd.cut(fire_df_ready['TOTALACRES'], bins=bins, labels=labels, right=True)
```
